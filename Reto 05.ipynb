# Problema 1
Dado la figura de la imagen, desarrolle:

* Una función matemática para calcular el volumen y el área superficial.
* Cree dos funciones en python para calcular los valores antes establecidos, al ingresar por teclado r1, r2 y h.
* Revise como utilizar el valor de pi usando import math y math.pi

```python
import math # Se importa la libreria de la cual se podra usar el pi posteriormente
pi=math.pi
sqrt=math.sqrt
def calcular_volumen(r1:float, r2:float, h:float):
 volumen_esfera = (4 / 3) * pi * (r1 ** 3)
 volumen_cono = (1 / 3) * pi * h * (r1 ** 2 + (r1 * r2) + r2 ** 2)
 return volumen_esfera, volumen_cono

def calcular_area_superficial(r1, r2, h):
  area_esfera = 4 * pi * (r1 ** 2)
  generatriz = sqrt(h ** 2 + (r1 - r2) ** 2)
  area_cono_lateral = pi * (r1 + r2) * generatriz
  area_base_cono = pi * r1 ** 2
  return area_esfera + area_cono_lateral + area_base_cono

if __name__ == "__main__":
    r1 = float(input("Ingrese el radio de la esfera: "))
    r2 = float(input("Ingrese el radio del cono: "))
    h = float(input("Ingresa la altura del cono: "))
    
    volumen_es, volumen_co = calcular_volumen(r1, r2, h)
    area_superficial = calcular_area_superficial(r1, r2, h)

print("El volumen de la esfera es", volumen_es)
print("El volumen del cono es", volumen_co)
print("El área superficial total es",area_superficial )
```

* Ingrese el radio de la esfera: 18
* Ingrese el radio del cono: 12
* Ingresa la altura del cono: 20
* El volumen de la esfera es 24429.02447431423
* El volumen del cono es 14325.662500369455
* El área superficial total es 7057.331512592609

# Problema 2
Dado la figura de la imagen, desarrolle:

* Una función matemática para calcular el área y el perimetro.
* Cree dos funciones en python para calcular los valores antes establecidos, al ingresar por teclado r, a y b.
* Revise como utilizar el valor de pi usando import math y math.pi

```python
import math # Se importa la libreria de la cual se podra usar el pi posteriormente
pi=math.pi
def area (a:float,b:float,r:float)->float:
    area_total=(a*b+2*pi*r**2)
    return area_total
def perimetro (a:float,b:float,r:float)->float:
  perimetro_total=2*a+2*b+2*(2*pi*r)
  return perimetro_total
if __name__=="__main__":
  a=float (input("Ingrese la altura del rectangulo en centimentros: "))
  b=float (input("Ingrese la base del rectangulo en centimentros: "))
  r=float (input("Ingrese el radio de los circulos en centimentros: "))


  area_final= area (a,b,r)
  perimetro_final= perimetro (a,b,r)
  print ("El area total de la figura es",area_final,"cm²" )
  print ("El perimetro total de la figura es",perimetro_final,"cm")
```

* Ingrese la altura del rectangulo: 5
* Ingrese la base del rectangulo: 18
* Ingrese el radio de los circulos: 3
* El area total de la figura es 146.54866776461628 cm²
* El perimetro total de la figura es 83.69911184307752 cm

# Problema 3
Diseñe una función que calcule la cantidad de carne de aves en kilos si se tienen N gallinas, M gallos y K pollitos cada uno pesando 6 kilos, 7 kilos y 1 kilo respectivamente.

```python
kilos_gallinas=6
kilos_gallos=7
kilos_pollitos=1
 

def kilos_totales (N,M,K):
    kilos_t= N*kilos_gallinas + M*kilos_gallos + K*kilos_pollitos
    return kilos_t

if __name__=="__main__":
    N=int(input("Ingrese en número de gallinas: "))
    M=int(input("Ingrese en número de gallos: "))
    K=int(input("Ingrese en número de pollitos: "))

    peso_total=kilos_totales (N,M,K)
    print ("La cantidad de carne de aves es",peso_total,"Kg")
```

* Ingrese en número de gallinas: 15
* Ingrese en número de gallos: 18
* Ingrese en número de pollitos: 20
* La cantidad de carne de aves es 236 Kg

# Problema 4
Haga un programa que utilice una función para calcular el valor de un préstamo C usando interés compuesto del i por n meses.

```python
def valor_prestamo(C, i, n):
    valor_total = C * (1 + (i/100)) ** (n/12)
    return valor_total

if __name__ == "__main__":
    C = float(input("Ingrese la cantidad del préstamo: "))
    i = float(input("Ingrese la tasa de interés anual: "))
    n = int(input("Ingrese en cuantos meses quiere el prestamo: "))

    valor_f = valor_prestamo(C, i, n)
    print("El valor final del préstamo es:",valor_f)
```

* Ingrese la tasa de interés anual: 13
* Ingrese en cuantos meses quiere el prestamo: 9

# Problema 5
Escriba un programa que pida 5 números reales y calcule las siguientes operaciones usando una función para cada una:
* El promedio
* El promedio multiplicativo (multilplica todos y luego calcula la raíz de la cantidad de operandos)
* La potencia del mayor número elevado al menor número
* La raíz cúbica del menor número

```python
import math # Se importa la libreria 
cbrt=math.cbrt
def promedio(n1: float, n2:float, n3:float, n4:float, n5: float):
  prom = (n1+n2+n3+n4+n5)/5
  return prom

def promedio_multiplicativo(n1: float, n2:float, n3:float, n4:float, n5: float):
 promedio_mul = (n1*n2*n3*n4*n5)**(1/5)
 return promedio_mul

def inicio_potencia (n1: float, n2:float, n3:float, n4:float, n5: float):
    mayor = max (n1,n2,n3,n4,n5)
    menor = min (n1,n2,n3,n4,n5)
    total_potencia = mayor ** menor
    return total_potencia

def inicio_raiz (n1: float, n2:float, n3:float, n4:float, n5: float):
    menor_r = min (n1,n2,n3,n4,n5)
    total_raiz = cbrt (menor_r)

    return total_raiz
if __name__ == "__main__":
 n1 = float(input("Ingrese el primer numero: "))
 n2 = float(input("Ingrese el segundo numero: "))
 n3 = float(input("Ingrese el tercer numero: "))
 n4 = float(input("Ingrese el cuarto numero: "))
 n5 = float(input("Ingrese el quinto numero: "))

promedio_final = promedio(n1,n2,n3,n4,n5)
prom_final = promedio_multiplicativo(n1,n2,n3,n4,n5)
potencia_final = inicio_potencia(n1,n2,n3,n4,n5)
raiz_final = inicio_raiz(n1,n2,n3,n4,n5)

print("El promedio es", promedio_final)
print("El promedio multiplicativo es", prom_final)
print("La potencia del mayor número elevado al menor número es ",potencia_final)
print("La raíz cúbica del menor número es ", raiz_final)
```

* Ingrese el primer numero: 15
* Ingrese el segundo numero: 5
* Ingrese el tercer numero: 20
* Ingrese el cuarto numero: 19
* Ingrese el quinto numero: 4
* El promedio es 12.6
* El promedio multiplicativo es 10.265520397411134
* La potencia del mayor número elevado al menor número es  160000.0
* La raíz cúbica del menor número es  1.5874010519681994

# ¿Que es un pip?
* Pip es un gestor de paquetes de Python usado para instalar, actualizar y eliminar software.
* Viene incluido por defecto en Python 2.7.9 y posteriores, así como en Python 3.4 y versiones superiores.
* El alias para pip en Python 3 es pip3, permitiendo a los usuarios trabajar con paquetes en esa versión.
* Los usuarios pueden instalar paquetes desde archivos locales o repositorios externos.
